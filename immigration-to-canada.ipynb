Create a line plot on the above dataset from 1980-2010

```python
# step 2: plot data
canadaimmigration.plot(kind='line', figsize=(10, 6))

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Immigrants from other countries to Canada from 1980 to 2010') # add title to the plot

plt.show()
```

Total count on immigrants moving into canada.

```python
import matplotlib.pyplot as plt
canadaimmigration.Total.value_counts().nlargest(10).plot(kind='bar', figsize=(15,5))
plt.title("Number of Immigrants")
plt.ylabel('Count of Immigrants')
plt.xlabel('Total');
```
